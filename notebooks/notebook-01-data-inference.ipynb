```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
import os
import warnings
warnings.filterwarnings('ignore')

DATAFOLDER = os.path.join('..', 'data', 'csv')
DATAS = {'FOURIER': 'v000_SCIG_SC_SENSORA_FOURIER', 'HOS': 'v000_SCIG_SC_SENSORA_HOS', 'SCM': 'v000_SCIG_SC_SENSORA_SCM'}
# FEATURES_NAMES = {'FOURIER': ['fx0d5_R', 'fx1d5_R', 'fx2d5_R', 'fx3_R', 'fx5_R', 'fx7_R', 'Freq_Gen', 'CC_bus'],
#                   'HOS': ['Skewness_R', 'Kurtosis_R', 'Variance_R', 'RMS_R', 'Freq_Gen', 'CC_bus'],
#                   'SCM': ['scm_COR_R', 'scm_IDM_R', 'scm_ENT_R', 'scm_CSD_R', 'scm_CSR_R', 'Freq_Gen', 'CC_bus']}
SEED = 6969
TEST_SIZE = 0.1
```

# Load data

```python
data = pd.read_csv(os.path.join(DATAFOLDER, DATAS['SCM'] + '.csv'))
data.sort_values(by='Class', inplace=True)
data.head()
```

```python

for feature_set in DATAS.keys():
    data = pd.read_csv(os.path.join(DATAFOLDER, DATAS[feature_set] + '.csv'))

    labels = data['Class']
    features = data

    X = features.values
    y = labels.values

    sss = StratifiedShuffleSplit(n_splits=1, test_size=TEST_SIZE, random_state=SEED)
    for train_index, test_index in sss.split(features, labels):
        print("TRAIN size:", len(train_index), "TEST size:", len(test_index))
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        pd.DataFrame(X_train, columns=features.columns).to_csv(os.path.join(DATAFOLDER, DATAS[feature_set] + '_chunk_{}.csv'.format(int(TEST_SIZE*100))), index=False)
        pd.DataFrame(y_train, columns=['Class']).to_csv(os.path.join(DATAFOLDER, DATAS[feature_set] + '_labels_chunk_{}.csv'.format(int(TEST_SIZE*100))), index=False)

        pd.DataFrame(X_test, columns=features.columns).to_csv(os.path.join(DATAFOLDER, DATAS[feature_set] + '_chunk_{}.csv'.format(int(TEST_SIZE*100))), index=False)
        pd.DataFrame(y_test, columns=['Class']).to_csv(os.path.join(DATAFOLDER, DATAS[feature_set] + '_labels_chunk_{}.csv'.format(int(TEST_SIZE*100))), index=False)
```
